```python
from numpy import random
my_list=[]
sum=0.0

my_list=[i for i in range(1,14) for j in range(1,5)]

for i in my_list:
    sum=sum+i
#z=1
def run_function(games=1): #function which has both deal and play functions defined within
    z=1
    for number in range(1,games+1):
        #print('\n GAME NO : {}'.format(number))

        def deal(): #Exercise 2
            random.shuffle(my_list)
            #print(my_list)

            player_a=tuple(my_list[:int(len(my_list)/2)])

            player_b=tuple(my_list[int(len(my_list)/2):])

            return player_a,player_b

        tuple_a,tuple_b=deal()

        sum_a=0
        sum_b=0
        count=0
        my_dict={}
        return_list=[]

        number_a=len(tuple_a)
        for i in tuple_a:
            sum_a=sum_a+i
        #print('no of cards with player a is {} and total initial value is {}'.format(number_a,sum_a))

        number_b=len(tuple_b)
        for i in tuple_b:
            sum_b=sum_b+i
        #print('no of cards with player b is {} and total initial value is {}'.format(number_b,sum_b))
        list_a=list(tuple_a)
        list_a.reverse()   # converted to list to use the pop function as tuples are immutable and reversed because pop eliminates the last element of a list

        list_b=list(tuple_b)
        list_b.reverse()


    # STEP 3.2 and 3.3 excuted below


        def play(): #game of ADDITION WAR is being played within this
            total_a=0
            total_b=0

            round=1
            discard_a=[] #this will have the final cards in hand of player A after a game is complete
            int_a=[] # if a game round is TIE , this is an intermediate array where the cards get stored
            discard_b=[] # this will have final cards in hand of player B after a game is complete
            int_b=[]
            count=0
            points_won=[]
            while list_a and list_b: #terminating condition specified here
                #print('round number is {}'.format(round))
                flag=0 
                a_1=list_a.pop() #select card 1 for player A
                a_2=list_a.pop() #select card 2 for player A
                b_1=list_b.pop() #similar for player B
                b_2=list_b.pop()
                value_a=a_1+a_2 #value of two card points computed
                value_b=b_1+b_2


                if value_a==value_b: #same card points for both players 
                    int_a.append(a_1)
                    int_a.append(a_2)
                    int_b.append(b_1)
                    int_b.append(b_2)
                    count+=1  # count is the variable to count no of ties within each game


                elif value_a>value_b: #player A has a higher card point than player B
                    sum_points_won=0
                    total_int_list=int_a + int_b
                    int_a.clear()
                    int_b.clear()
                    discard_a=discard_a+total_int_list
                    discard_a.append(a_1) #values get added to player A's final cards array 
                    discard_a.append(a_2)
                    discard_a.append(b_1)
                    discard_a.append(b_2)
                    for elem in total_int_list:
                        sum_points_won+=elem
                    sum_points_won+=a_1+a_2+b_1+b_2
                    points_won.append(sum_points_won) # array keeps track of total won by a player in each round

                elif value_a < value_b: #player B has a higher card point than player A
                    sum_points_won=0
                    total_int_list=int_a + int_b
                    int_a.clear()
                    int_b.clear()
                    discard_b=discard_b+total_int_list
                    discard_b.append(a_1) #values get added to player B's final cards array
                    discard_b.append(a_2)
                    discard_b.append(b_1)
                    discard_b.append(b_2)
                    for elem in total_int_list:
                        sum_points_won+=elem
                    sum_points_won+=a_1+a_2+b_1+b_2
                    points_won.append(sum_points_won)
                #following statement is commented as not needed in the output 
                #print(" card points with player A after round {} is {} \n and card points with player B after round {} is {}\n".format(round,discard_a,round,discard_b))
                round+=1
            #following statements are commented as not needed in output of HW3

            #print('final card point array of player A is : {}'.format(discard_a))
            #print('final card point array of player B is :{}'.format(discard_b))
            #print('number of ties in an entire game is: {}'.format(count)) # Step 1.1 HW3
            #print('largest pot amount won by any player in the entire game:{}'.format(max(points_won))) #largest total won during a round Step 1.2 HW3


            for i in discard_a:
                total_a=total_a+ i #computing total values to make the decision of who won 
            for j in discard_b:
                total_b=total_b+ j
            #not needed in the output as 1000 games have to be simulated
            #print('final totals for player 1 and player 2 at the end of game is {} and {}'.format(total_a,total_b))
            #appending all the required values to a dictionary which is then returned by the play() function
            my_dict['initial_values_p1']=str(sum_a)
            my_dict['initial_values_p2']=str(sum_b)
            my_dict['final_values_p1']=str(total_a)
            my_dict['final_values_p2']=str(total_b)
            my_dict['no_of_ties']=str(count)
            
            my_dict['largest_pot_win']=str(max(points_won))
            #not needed in output
            #print(my_dict)
            
            return my_dict

        dic_to_df = play()
        
        
        #not needed in output
        #print('\narray which includes all winning card point values : {}'.format(win_array))
        #print('\narray which includes initial card point value for corresponding winners: {}'.format(initial_value_array))
        #print('\narray for the loser values :{}'.format(lose_array))
        # EXERCISE 2 - named dataframe as sachi as it is a name of my loved one ;*
        if z==1:
            sachi=pd.DataFrame(list(dic_to_df.values()),index=['init_p1','init_p2','fin_p1','fin_p2','ties','large_pot'],columns=['obs_number:1'])
        else:
            sachi.insert(number-1,'obs_number:{}'.format(number),[dic_to_df['initial_values_p1'],dic_to_df['initial_values_p2'],dic_to_df['final_values_p1'],dic_to_df['final_values_p2'],dic_to_df['no_of_ties'],dic_to_df['largest_pot_win']])
            
        z+=1
        
    sachi_t=sachi.transpose() # transposed to swap the columns with rows and now the df name changed to sachi_t
    # Exercise 3 Step 3.1
    sachi_t['outcome']=""
    
    print('\n the dataframe is : \n')
    print(sachi_t.head(10))
    print(sachi_t.tail(10))
    for label,row_data in sachi_t.iterrows():
        if int(sachi_t.at[label,'fin_p1'])>int(sachi_t.at[label,'fin_p2']):
            row_data['outcome'] = 'Player 1 wins'
        elif int(sachi_t.at[label,'fin_p1'])<int(sachi_t.at[label,'fin_p2']):
            row_data['outcome'] = 'Player 2 wins'
        else :
            row_data['outcome'] ='Tie game!'
            
    # Exercise 3 Step 3.2
    
    sachi_t['winner_total']=np.nan
    
        
    sachi_t.loc[sachi_t['fin_p1'].astype(int) > sachi_t['fin_p2'].astype(int), 'winner_total'] = sachi_t.loc[sachi_t['fin_p1'].astype(int) > sachi_t['fin_p2'].astype(int), 'fin_p1']
    sachi_t.loc[sachi_t['fin_p1'].astype(int) < sachi_t['fin_p2'].astype(int), 'winner_total'] = sachi_t.loc[sachi_t['fin_p1'].astype(int) < sachi_t['fin_p2'].astype(int), 'fin_p2']
    sachi_t.loc[sachi_t['fin_p1'].astype(int) == sachi_t['fin_p2'].astype(int), 'winner_total'] = sachi_t.loc[sachi_t['fin_p1'].astype(int) == sachi_t['fin_p2'].astype(int), 'fin_p2']
    count_p1=0
    count_p2=0
    count_tie=0
    for label,row_data in sachi_t.iterrows():
        
        if (row_data['outcome'] == 'Player 1 wins'):
            count_p1+=1
        elif (row_data['outcome'] == 'Player 2 wins'):
            count_p2+=1
        else :
            count_tie+=1
    print('\n df with two new columns:')
    print(sachi_t.head(10)) 
    print('\n Frequency of A winning is {} , B winning is {}, tie game is {}'.format(count_p1,count_p2,count_tie))
    print('\njust rows wherein game is a TIE :')
    print(sachi_t.loc[sachi_t['outcome'] == 'Tie game!'])
    print('\n')
    
    list_of_winning_values=list(map(float,sachi_t["winner_total"].tolist()))
    
    #print('the column winner_total has been extracted :{}'.format(list_of_winning_values))
    print('\n Step 3.4 Descriptive Statistics:')
    print('std dev is {}'.format(statistics.stdev(list_of_winning_values)))
    print('minimum is {}'.format(min(list_of_winning_values)))
    print('maximum is {}'.format(max(list_of_winning_values)))
    print('mean is {}'.format(statistics.mean(list_of_winning_values)))
    print('\n df with sorted winner_total values from large to small:')
    print(sachi_t.sort_values(by='winner_total',ascending=False).head(3))
    #following code is for step 4.2 to calculate mean max round total as a function of tie
    list_of_tie_games=np.array(list(map(int,sachi_t["ties"].tolist())))
    list_of_max_round_total= np.array(list(map(int,sachi_t["large_pot"].tolist())))
    num=(list_of_tie_games*list_of_max_round_total)
    final_num=0
    for i in num:
        final_num+=i
    deno= np.count_nonzero(list_of_tie_games)
    # Exercise 4 Step 4.2
    print('mean of max round total as a function of number of ties: {}'.format(final_num/deno))
    print('\nOBSERVATION- I observe that the mean will be close to that max round total which has more number of ties.')
```
